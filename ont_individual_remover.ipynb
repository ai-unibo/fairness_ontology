```python
from rdflib import Graph
import pandas as pd
from rdflib import URIRef,Literal
from rdflib.namespace import RDF,RDFS
```

```python
prefix="http://fairness_ontology.com/"
g = Graph()
g.parse("./complete-ontologies/fairness_ontology-3.ttl", format='turtle')
print(len(g))
#REMOVING INDIVIDUAL NOTIONS
not_query = """PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX : <http://fairness_ontology.com/> 
    PREFIX fmo: <https://purl.org/twc/fmo#> 
    SELECT ?individual
    WHERE {
    ?individual rdf:type ?subclass .
    ?subclass rdfs:subClassOf* fmo:fairness_notion .
    }"""
qres = g.query(not_query)
for row in qres:
    print(f"Removing NOTION {row.individual}")
    g.remove((row.individual, None, None))
    g.remove((None, None, row.individual))
```
